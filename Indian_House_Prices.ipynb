```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data= pd.read_csv("House Price India.csv")
```

```python
data.groupby("condition of the house")["Price"].mean().sort_values(ascending=False).plot(kind="bar")
plt.title("Condition of the house vs Price")
plt.ylabel("Mean Price")
plt.xlabel("House Condition")
plt.show()
```

```python
data.columns
```

```python
plt.figure(figsize=(15,8))
sns.heatmap(data.corr(), annot=True, cmap='YlGnBu')
```

```python
x=data[["number of bedrooms",'number of bathrooms','living area','number of floors','waterfront present',
        'grade of the house','Area of the house(excluding basement)','Area of the basement','number of views',
        'condition of the house','Built Year', 'Number of schools nearby', 'Distance from the airport',
       'living_area_renov']]
y=data['Price']
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2)
```

```python
from sklearn.linear_model import LinearRegression

reg= LinearRegression()
reg.fit(x_train, y_train)
```

```python
reg.score(x_test,y_test)
```

```python
from sklearn.ensemble import RandomForestRegressor

forest = RandomForestRegressor()

forest.fit(x_train,y_train)
forest.score(x_test,y_test)
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
from sklearn.tree import DecisionTreeRegressor
param_grid={
    "criterion":['mse','friedman_mse','mae'],
    "splitter" :["best","random"],
    "max_depth":[None,10,20,30,40,50],
    "min_samples_split":[2,5,10],
    "min_samples_leaf":[1,2,4]
    
}
```

```python
tree_model= DecisionTreeRegressor()

grid_tree= GridSearchCV(estimator=tree_model, param_grid=param_grid)
```

```python
grid_tree.fit(x_train,y_train)
```

```python
grid_tree.score(x_test,y_test)
```
